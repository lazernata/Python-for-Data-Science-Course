```python
import pandas as pd
import requests
df_ventas = pd.read_excel(r'C:\Users\nlazareva\Downloads\Dataset-Actividad-Practica-1\CASO PRÁCTICO 1\Ventas.xlsx')
df_clientes = pd.read_excel(r'C:\Users\nlazareva\Downloads\Dataset-Actividad-Practica-1\CASO PRÁCTICO 1\Clientes.xlsx')
df_productos = pd.read_excel(r'C:\Users\nlazareva\Downloads\Dataset-Actividad-Practica-1\CASO PRÁCTICO 1\Productos.xlsx')
```

```python
r = requests.get('https://www.worldometers.info/gdp/gdp-by-country/')
df_list = pd.read_html(r.text)[0]
df_pob = df_list[['Country', 'Population  (2022)']]
df_pib = df_list[['Country', 'GDP  (nominal, 2022)']]
df_rpc = df_list[['Country', 'GDP per capita']]
```

```python
# 1
lista = [df_ventas, df_clientes, df_productos, df_pob, df_pib, df_rpc]
names = ['df_ventas', 'df_clientes', 'df_productos', 'df_pob', 'df_pib', 'df_rpc']

for name, df in zip(names, lista):
    print(name, df.shape)
```

```python
# 2
for name, df in zip(names, lista):
    if df.isnull().any().any():
        print(f"There are NaN values in the {name}.")
    else:
        print(f"There are no NaN values in the {name}.")
```

```python
# 9
masde13 = df_ventas1[df_ventas1['Cantidad'] > 13]
df_ventas1 = df_ventas1.drop(masde13.index)
print('El nuevo dataframe de ventas tiene las siguientes dimensiones: ', df_ventas1.shape)
```

```python
# 10
diferencias = (df_ventas1['Fecha envío'] - df_ventas1['Fecha compra']).dropna()
mediana_diferencias = diferencias.median()

df_ventas1['Fecha envío'] = df_ventas1['Fecha envío'].fillna(df_ventas1['Fecha compra'] + mediana_diferencias)

df_ventas1
```

```python
# 11
df_comb = pd.merge(df_clientes, df_ventas1, on='ID Cliente', how='inner')
df_comb = pd.merge(df_comb, df_productos, on='ID Producto', how='inner')
df_comb
```

```python
df_comb.groupby('País')['Cantidad'].sum() # EEUU tienen mayor número de ventas
```

```python
# 11
df_comb.groupby(df_comb['Fecha compra'].dt.year)['Cantidad'].sum() # el año 2021 tiene mayor número de ventas
```

```python
df_comb.head()
```

```python
# 12
df_comb['Beneficio'] = (df_comb['Precio unitario'] * df_comb['Cantidad']) - df_comb['Coste Producción']
df_comb.head()
```

```python
df_comb.groupby('Categoría')['Beneficio'].sum() # La categoría Tecnología genera mayores beneficios
```

```python
# 14
df_rpc = df_rpc.rename(columns={'Country': 'País'})
df_rpc['País']=df_rpc['País'].str.lstrip()
df_rpc.head()
```

```python
df_comb = pd.merge(df_comb, df_rpc, on='País', how='left')
df_comb.head()
```

```python
df_comb.corr(numeric_only = True) # La correlación entre GDP per cpita y Beneficio es prácticamente nula (=-0.000990)
```

```python
# 15
df_comb.corr(numeric_only = True) # Sí, existe una correlacción moderada entre Cantidad y Beneficio (=0.345548)
```

```python
# 16
df_pib = df_pib.rename(columns={'Country': 'País'})
df_pib['País']=df_pib['País'].str.lstrip()
df_pib.head()
```

```python
df_pob = df_pob.rename(columns={'Country': 'País'})
df_pob['País']=df_pob['País'].str.lstrip()
df_pob.head()
```

```python
df_comb = pd.merge(df_comb, df_pib, on='País', how='left')
df_comb = pd.merge(df_comb, df_pob, on='País', how='left')
df_comb.head()
```

```python
df_comb.corr(numeric_only = True)
# Existe una correlacción fuerte positiva entre la Población y GDP per capita y GDP nominal (=0.948383 y =0.999636), y también
# entre GDP per capita y GDP nominal (=0.940764). 
# Es decir, comforme va aumentando la población, crece GDP per capita y GDP nominal y al revés. 
```

```python
# 17
df_comb['Tiempo Envío'] = df_comb['Fecha envío'] - df_comb['Fecha compra']
df_comb.head()
df_comb.groupby('País')['Tiempo Envío'].mean() # En Alemania y Rumanía el tiempo promedio es mayor
```

```python
# 18
beneficios_por_cat_pais = df_comb.groupby(['País','Categoría'])['Beneficio'].sum()
beneficios_por_cat_pais.head()
```

```python
beneficios_por_cat_pais_df = pd.DataFrame(beneficios_por_cat_pais)
beneficios_por_cat_pais_df.to_excel('Reporte Beneficios.xlsx', sheet_name='Beneficios_cat_pais', index=True)
beneficios_por_cat_pais_df
```
