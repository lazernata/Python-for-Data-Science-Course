```python
df= pd.read_csv(r"C:\Users\nlazareva\Downloads\SP500_data_2019_2020.csv", encoding = 'ISO-8859-1', delimiter = ',')
df.head()
```

```python
window_size = 60
windows = df['Close'].rolling(window_size)
moving_averages = windows.mean()
moving_averages_list = moving_averages.tolist()
```

```python
plt.plot(df['Date'], df['Close'], label = 'Close per dia')
plt.plot(df['Date'], moving_averages_list, label = 'Media móvil per dia')
plt.legend()
```
