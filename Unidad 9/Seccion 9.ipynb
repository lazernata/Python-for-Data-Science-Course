```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\nlazareva\Downloads\SP500_data_2019_2020_1.csv", encoding = 'ISO-8859-1', delimiter = ',',
                parse_dates = True, index_col = 'Date')
df.head()
```
