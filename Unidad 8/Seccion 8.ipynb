```python
import pandas as pd
import seaborn as sns
from scipy import stats
```

```python
total_historico = pd.read_excel(r"C:\Users\nlazareva\Downloads\Datos_poblacin_activa.xlsx", sheet_name = 'Total Hist√≥rico')
total_historico.head()
```

```python
sns.regplot(data = total_historico, x = 'Periodo num_int', y = 'Total Activos', scatter = True, color = 'blue', label = 'order 1')
```

```python
pendiente, intercepto, r_value, p_value, std_err = stats.linregress(total_historico['Periodo num_int'], total_historico['Total Activos'])
```

```python
sns.regplot(data = total_historico, x = 'Periodo num_int', y = 'Total Activos', scatter = True, 
            color = 'blue', label = 'order 1', line_kws = {'label':"y={0:.1f}x+{1:.1f}".format(pendiente,intercepto)})
```
